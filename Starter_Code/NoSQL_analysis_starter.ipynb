### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
# Print the results
```

```python
# Define the latitude and longitude values for the center point
center_latitude = 51.490142
center_longitude = 0.083840

# Define the latitude and longitude range
latitude_range = 0.01
longitude_range = 0.01

# Calculate the latitude and longitude bounds
min_latitude = center_latitude - latitude_range
max_latitude = center_latitude + latitude_range
min_longitude = center_longitude - longitude_range
max_longitude = center_longitude + longitude_range

# Find the establishments within the specified range
establishments_within_range = establishments.find({
    "geocode.latitude": {"$gte": min_latitude, "$lte": max_latitude},
    "geocode.longitude": {"$gte": min_longitude, "$lte": max_longitude},
    "RatingValue": 5
}).sort("scores.Hygiene")

# Convert the result to a Pandas DataFrame
df_within_range = pd.DataFrame(establishments_within_range)

# Display the DataFrame
df_within_range

```

```python
# Convert result to Pandas DataFrame
import pandas as pd

# Convert the result to a Pandas DataFrame
df_within_range = pd.DataFrame(establishments_within_range)

# Display the DataFrame
print(df_within_range)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Print the number of documents in the result

# Print the first 10 results


# Create a pipeline
pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}}
]

# Execute the pipeline
result = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of documents in the result:", len(result))

# Print the first 10 results
print("First 10 results:")
for idx, doc in enumerate(result[:10], start=1):
    print(f"{idx}. Local Authority: {doc['_id']}, Count: {doc['count']}")

```

```python
# Convert the result to a Pandas DataFrame

# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame


import pandas as pd

# Convert the result to a Pandas DataFrame
df = pd.DataFrame(result)

# Display the number of rows in the DataFrame
num_rows = df.shape[0]
print("Number of rows in the DataFrame:", num_rows)

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
df.head(10)
```
